```python
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from PIL import Image
%matplotlib inline
```

```python
text = open("All_I_Want_For_Christmas.txt").read()
tree_mask = np.array(Image.open("maska.png"))
```

```python
wc = WordCloud(background_color="white", max_words=2000, mask=tree_mask, contour_width=3, contour_color='steelblue')
wc.generate(text)
```

```python
wc.to_file("christmas_tree.png")
```

```python
plt.figure(figsize = (10,12))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
```
